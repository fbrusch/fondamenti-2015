```python
def complementa1(n):
    return [1-x for x in n]
```

```python
x = [0,1,0,1,1]
```

```python
complementa1(x)
```

```python
def half_adder(b1, b2):
    if b1 == 0 and b2 == 0: return [0, 0]
    if b1 == 1 and b2 == 0: return [0, 1]
    if b1 == 0 and b2 == 1: return [0, 1]
    return [1, 0]
```

```python
half_adder(1, 0)
```

```python
def full_adder(b1, b2, b3):
    s = b1 + b2 + b3
    if s == 0: return [0, 0]
    if s == 1: return [0, 1]
    if s == 2: return [1, 0]
    return [1, 1]
```

```python
full_adder(1, 1, 1)
```

```python
full_adder(0, 1, 1)
```

[1, 0] -> [0, 1, 1]

[0, 0] -> [0, 0, 1]

[0, 1] -> [0, 1, 0]

[1, 1] -> [1, 0, 0]

```python
def somma_1(n):
    s = [0, 0, 0]
    s_tmp = half_adder(n[1], 1)
    s[2] = s_tmp[1]
    s_tmp2 = half_adder(n[0], s_tmp[0])
    s[1] = s_tmp2[1]
    s[0] = s_tmp2[0]
    return s
```

```python
somma_1([0, 1])
```

```python
somma_1([1, 0])
```

```python
somma_1([1, 1])
```
